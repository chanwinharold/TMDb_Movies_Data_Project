# Visualisation des donn√©es

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

data = pd.read_csv("../data/data_cleaned.csv")
```

### Quels sont les genres avec le plus de revenues moyen ?

```python
df_genre = data.groupby("genres")["revenue_adj"].mean()
plt.xticks(rotation=90)
sns.barplot(data=pd.DataFrame(df_genre), x="genres", y="revenue_adj", hue="genres", palette="Set2")
plt.show()
```

### Quels sont les genres avec le plus de budget moyen ?

```python
df_genre = data.groupby("genres")["budget_adj"].mean()
plt.xticks(rotation=90)
sns.barplot(data=pd.DataFrame(df_genre), x="genres", y="budget_adj", hue="genres", palette="Set2")
plt.show()
```

```python
for x in ["budget_adj", "vote_average"]:
    sns.scatterplot(data=data, x=x, y="revenue_adj")
    plt.show()
```

D'apr√®s ces nuages de points, on voit bien que les films avec le plus de revenues ont leurs votes moyens qui tourne autour de "7" et ils ont un √©galement plus de budget. Comme quoi un bon film, c'est tout d'abord un bon budget.
